```python
#Importing an Image:
import matplotlib.pyplot as plt
import imageio.v2 as imageio

path = "img.png"
pic = imageio.imread(path)

plt.figure(figsize = (10,10))
plt.title("Image")
plt.imshow(pic)
plt.show()
```

```python
# Understanding the underlying data: Red Intensity
red =imageio.imread(path)

red[:, :, 0] = 255
plt.figure(figsize = (10,10))
plt.imshow(red)
plt.title("Red Intensity")
plt.show()
```

```python
# Understanding the underlying data: Green Intensity
green = imageio.imread(path)

green[:, :, 1] = 255
plt.figure(figsize = (10,10))
plt.imshow(green)
plt.title("Greem Intensity")
plt.show()
```

```python
#Understanding the underlying data: Blue Intensity
blue= imageio.imread(path)

blue[: , : , 2] = 255 
plt.figure( figsize = (10,10))
plt.imshow(blue)
plt.title("Blue Intensity")
plt.show()
```

```python
# Compressing the image to a 2-D matrix by Gray-scaling
import numpy as np
gray = lambda rgb : np.dot(rgb[... , :3] , [0.299 , 0.587, 0.114]) 
gray = gray(pic)

plt.figure(figsize = (10,10))
plt.title("Gray Scale")
plt.imshow(gray, cmap = plt.get_cmap(name = 'gray'))
plt.show()

#Printing shapes
print("Colored image shape:", pic.shape)
print("Grayscale image shape:", gray.shape)
```

```python
# To binarize this Gray scale image: segmentation of pixels into foreground and background. 
from skimage.filters import threshold_otsu
thresh = threshold_otsu(gray)
binary = gray > thresh

plt.figure( figsize = (10,10))
plt.title("Binary Image")
plt.imshow(binary, cmap = plt.get_cmap(name = 'gray'))
plt.show()
```

```python
# Blurring Image
from skimage.filters import gaussian

blurred = gaussian(gray,sigma=20)
plt.figure(figsize = (10,10))
plt.title("Blurred Image")
plt.imshow(blurred, cmap = plt.get_cmap(name = 'gray'))
plt.show()
```

```python
# Plotting histogram to show how frequently various colour values occur in the image.
# To calculate frequency of pixels in range 0-255

import cv2
histr = cv2.calcHist([pic],[0],None,[256],[0,256])
plt.plot(histr)
plt.xlabel("Pixel Value")
plt.ylabel("Number/Frequency")

plt.show()
```
