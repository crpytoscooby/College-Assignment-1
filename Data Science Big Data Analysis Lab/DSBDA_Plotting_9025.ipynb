```python
import numpy as np 
import pandas as pd
```

```python
data = pd.read_csv("/content/drive/MyDrive/TE Sem 2/Heart.csv")
```

```python
data.head
```

```python
# find null values

data.isnull().sum()
```

```python
# fill those null values with mean of the column
dataset=data.fillna(data.mean())
dataset
```

```python
# checking if still their are any null values
dataset.isnull().sum()
```

```python
heart_data = pd.DataFrame(dataset, columns = ['Age', 'Sex', 'ChestPain', 'RestBP', 'Chol', 'Fbs', 'MaxHR', 'ExAng', 'Thal', 'AHD', 'Slope', 'Oldpeak', 'RestECG'])
heart_data
```
