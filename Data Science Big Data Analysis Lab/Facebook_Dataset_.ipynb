```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv("dataset_Facebook.csv", sep=';')
```

```python
df.head()
```

```python
df.info()
```

```python
df['share'] = df['share'].fillna(df['share'].mean)
df['Paid'] = df['Paid'].fillna(1.0)
df['like'] = df['like'].fillna(df['like'].mean)
```

```python
df.isnull().sum()
```

Subsets of rows and columns

```python
row1 = df.iloc[:5]
row1
```

```python
row2 = df.tail(5)
row2
```

```python
col1 = df[['Type','Category']]
col1
```

```python
col2 = df[['comment','like']]
col2
```

Merge Data



```python
merge_rows = pd.concat([row1, row2], axis=0)
merge_rows
```

```python
merge_cols = pd.concat([col1, col2], axis=1)
merge_cols
```

Sort Data in Ascending and Descending order

```python
ascending = df.sort_values('comment', ascending=True)
ascending
```

```python
descending = df.sort_values('comment', ascending=False)
descending
```

```python
transpose_data = df.transpose()
transpose_data
```
